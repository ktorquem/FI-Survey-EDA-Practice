## Preprocessing


#### Change Yes/No values to 1/0 and Male/Female to 0/1 (Make Other = Male)

```python
d[d['Future Age at FI'] == 'Old'] = nan
d[d['Future Age at FI'] == 'Idk'] = nan
d[d['Future Year at FI'] == 'Idk'] = nan

for col,row in d.iteritems():
    if isinstance(row[0], str):
        d.loc[d[col] == 'No', col] = 0
        d.loc[d[col] == 'Yes', col] = 1
        d.loc[d[col] == 'Male', col] = 0
        d.loc[d[col] == 'Female', col] = 1
        d.loc[d[col] == 'Other', col] = 0
        d[col] = pd.to_numeric(d[col])
```

```python
d = d[pd.notnull(d.Gender)]
d = d[pd.notnull(d['Gross Income'])]
d = d[pd.notnull(d['Target SWR'])]
for x in d.columns:
    d[x].fillna(0, inplace=True)
```

```python
d = d[d['Student Loans'] < 450000]
d = d[d['SL APR'] < 20]
d.ix[307,'CC APR'] = 0  # Someone put 2016 here for APR but 0 CC Debt
d = d[d['Auto APR'] < 20] # More bad data
```
